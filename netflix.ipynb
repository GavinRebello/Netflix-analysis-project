```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\gavin\Downloads\netflix_titles.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
df[['director']].head()
```

```python
df["cast"]= df["cast"].fillna("Unknown")
df[["cast"]].head()
```

```python
df["country"]= df["country"].fillna("Other")
df[['country']]
```

```python
df['date_added']= df.apply(
    lambda row: f"{row['release_year']}-01-01" if pd.isnull(row['date_added']) else row['date_added'], axis=1
)
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
#pd.options.display.max_rows = None 
df[['date_added']]
```

```python
df['rating']= df['rating'].fillna(df['rating'].mode()[0])
df['rating']
pd.options.display.min_rows = None 
df['rating'].head(10)
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.duplicated().sum()
```

```python
df.drop(columns = ['description'] ,inplace=True)
df
```

```python
df["title"] = df["title"].str.strip()
df['title']
```

```python
df['cast']= df['cast'].apply(lambda x: ', '.join(x.split(', ')[:3]) if isinstance(x, str) else x)
df['cast']
```

```python
df['date_added']= df.apply(
    lambda row: f"{row['release_year']}-01-01" if pd.isnull(row['date_added']) else row['date_added'], axis=1
)
```

```python
df.to_csv('cleaned_netflix_data.csv', index=False)
```
